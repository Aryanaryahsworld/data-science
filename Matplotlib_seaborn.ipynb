```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
#Bar plot used in counting the number of passengers
sns.countplot(x='class', data=titanic)
plt.title('Passenger Class Count')
plt.show()
```

```python
#Age Distribution
sns.histplot(titanic['age'], kde=False, bins=10)
plt.title('Age Distribution of Passengers')
plt.show()
```

```python
#Gender distribution
gender_counts = titanic['sex'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.show()
```

```python
sns.boxplot(y='fare', data=titanic)
plt.title('Fare Distribution')
plt.show()
#Fare distribution box plot
```

```python
#age vs fare using scatter plot
sns.scatterplot(x='age', y='fare', data=titanic)
plt.title('Age vs. Fare')
plt.show()
```

```python
#Multivariate analysis Pair plot
sns.pairplot(titanic[['age', 'fare', 'class']], hue='class')
plt.show()
```

```python
sns.violinplot(x='class', y='age', data=titanic)
plt.title('Age Distribution by Passenger Class')
plt.show()
# Violet plot age distribution by class
```

```python
sns.stripplot(x='class', y='age', hue='sex', data=titanic)
plt.title('Class vs. Age with Gender')
plt.show()
#strip plot age distribution vs class
```

# **Advanced approach
**

```python
#survival rate by class gender hue
sns.catplot(x='class', hue='sex', col='survived', data=titanic, kind='count')
plt.subplots_adjust(top=0.8)
plt.suptitle('Survival Rate by Class and Gender')
plt.show()
```

```python
#Kernel Density estimation for age
sns.histplot(titanic['age'], kde=True, bins=20, color='green')
plt.title('Age Distribution with Kernel Density Estimation')
plt.show()
```

```python
#fair distribution by class gender hue using violin plot
sns.violinplot(x='class', y='fare', hue='sex', split=True, data=titanic)
plt.title('Fare Distribution by Class and Gender')
plt.show()

```

```python
#joinjt plot - age and fare with regression line
sns.jointplot(x='age', y='fare', data=titanic, kind='reg', color='purple')
plt.suptitle('Age vs. Fare with Regression Line', y=1.02)
plt.show()
```

```python
#swarm plot combined with violin plot age by class
plt.figure(figsize=(10, 6))
sns.violinplot(x='class', y='age', data=titanic, inner=None, color='lightgray')
sns.swarmplot(x='class', y='age', data=titanic, hue='sex')
plt.title('Age Distribution by Class with Swarm Plot Overlay')
plt.show()
```

```python
#pair plot advanved multivariate analysis
sns.pairplot(titanic.dropna(), vars=['age', 'fare', 'pclass'], hue='sex', diag_kind='kde', palette='Dark2')
plt.suptitle('Advanced Multivariate Analysis: Age, Fare, and Class', size=16, y=1.02)
plt.show()
```

```python
#heat map corelation matrix
corr = titanic.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
#facet grid mutliple histograms for age by class by gender
g = sns.FacetGrid(titanic, col='class', row='sex')
g.map(sns.histplot, 'age', bins=15, kde=False)
g.add_legend()
plt.show()
```
